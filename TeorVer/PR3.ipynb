```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import tan, exp, pi, sin

rmax = 6378137.0
rmin = 6356752.3142
e = 2*((rmax-rmin)/rmax) - ((rmax-rmin)/rmax)**2
```

```python
df = pd.read_csv('cities.csv', delimiter=',')
df = df[df['Population'] > 1000000]
df.head()
```

```python
df['DD Latitude'] = df['Longitude deg'] + df['Longitude min']/60 + df['Longitude sec']/3600 
df['DD Longitude'] = df['Latitude deg'] + df['Latitude min']/60 + df['Latitude sec']/3600
df.head()
```

```python
df['Xcor'] = df['DD Longitude'] * rmax
a = pi/4 + df['DD Latitude']/2
b = e*sin(df['DD Latitude'].to_list())
df['Ycor'] = rmax * exp(tan(a) * ((1 - b)/(1 + b))**(e/2))
df
```

```python
def get_pic(name):
    with open(name, 'rb') as pic:
        return plt.imread(pic)

fig = plt.figure()
pic = get_pic('world.jpg')

plt.imshow(pic)
plt.plot(df['Xcor'], df['Ycor'], 'r*')


plt.show()
```
