```python
import numpy as np
import matplotlib.pyplot as plt
m = 6
dots = np.array([[0, 1], [1, 1], [2, 25], [-1, 1], [-2, -23], [0.5, 0.90625]], dtype=float)


M = np.array([[dots[i][0]**j for j in range(m)] + [dots[i][1]] for i in range(m)])
print(*np.linalg.solve(M[:, :-1], M[:, -1]))
```

```python
def f(x):
    return 1 - x**3 + x**5
```

```python
x = np.linspace(-2, 2, 100)
y = f(x)
```

```python
ax = plt.gca()
ax.plot(x, y, 'b')
ax.spines['bottom'].set_position('center')
ax.spines['left'].set_position('center')
ax.semilogy()
```
