```python
def TestFun1(x, l, a, b):
    return (l*np.exp(l*x))/(np.exp(l*b) - np.exp(l*a))

def TestFun2(x, l):
    return 0.5 * abs(l)*np.exp(-1*abs(l*x))

def TestFun3(x, a, b):
    return (3*x**2)/((b**3) - (a**3))

def TestFun4(x, l):
    return TestFun2(l, x)

def TestFun5(x, a, b):
    return (2*x**2)/((b**3) - (a**3))

def TestFun6(x, a):
    return ((4)/(np.pi*a)) * np.arctan(x/a)

def TestFun7(x, a):
    return (2/(a*(1 + np.log(2)))) * np.log(x/a)

def TestFun8(x, l):
    return (1/l) * (np.sin((2*np.pi*x)/l))**2
```

```python
def integrate(f, x1, x2, acc, *args):
    xdata = np.arange(x1, x2, acc)
    ydata = f(xdata, *args) * acc
    res = sum(ydata)
    return res
```

```python
def get_shot(p):
    res = []
    r = random.random()
    while r < 1 - p:
        res.append(0)
        r = random.random()
    else:
        return len(res) + 1

res = []
for _ in range(1000):
    res.append(get_shot(0.2))
print(sum(res)/len(res))
```

```python
import numpy as np
from scipy.integrate import quad

def F(x):
    return 1 - np.exp(-x**2 / 2) if x >= 0 else 0

def f(x):
    return x * np.exp(-x**2 / 2) if x >= 0 else 0

# Нормализованная плотность g(x) пропорциональна производной G(x):
# g(x) = C * ∫₀¹ [f(x + u) - f(x - u)] * arctg(u) du
def g_unnormalized(x):
    integrand = lambda u: (f(x + u) - f(x - u)) * np.arctan(u)
    return 2 * quad(integrand, 0, 1, epsabs=1e-9)[0]

# Найдем нормировочную константу C:
a, b = 0, 5  # Достаточно широкий интервал
integral_g, _ = quad(g_unnormalized, a, b, epsabs=1e-9)
C = 1 / integral_g

# Нормированная плотность
def g(x):
    return C * g_unnormalized(x)

# Математическое ожидание E[X]
E_X, _ = quad(lambda x: x * g(x), a, b, epsabs=1e-9)

# Математическое ожидание квадрата: E[X²]
E_X2, _ = quad(lambda x: x**2 * g(x), a, b, epsabs=1e-9)

# Дисперсия: Var(X) = E[X²] - (E[X])²
variance = E_X**0.5

print(f"Дисперсия: {variance:.9f}")
```
