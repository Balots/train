```python
import matplotlib.pyplot as plt

def get_pic(name):
    with open(name, 'rb') as pic:
        return plt.imread(pic)

pic1, pic2, pic3, pic4 = map(get_pic, ['1.png', '2.png', '3.png', '4.png'])
```

```python
fig, p1 = plt.subplots()
p2 = fig.add_axes([0, -0.35, 0.5, 0.5])
p3 = fig.add_axes([0, -0.35, 1, 0.5])
p4 = fig.add_axes([0, -0.35, 1.5, 0.5])

p1.imshow(pic1)
p2.imshow(pic2)
p3.imshow(pic3)
p4.imshow(pic4)

p1.axis('off')
p2.axis('off')
p3.axis('off')
p4.axis('off')
```
