Набор данных представляет собой статистику параметров автомобилей на вторичном рынке в Молдавии. 

Набор включает ряд категориальных и численных значений, составляющих одну запись (строку). 

Каждый столбец в записи — это отдельный параметр. 

Среди указанных параметров приведен целевой для задачи предсказания непрерывного значения (регрессии) - цена автомобиля [$Price$]. 

 Также среди параметров есть целевой для задачи классификации - тип трансмиссии [$Transmission$]. 


# Import библиотек

```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns;
```

```python
from LinReg import LinearRegression
```

## Считываем файл в DataFrame



Метод .read_csv(path,delimiter) 

```python
df = pd.read_csv('data/cars_moldova_cat_num.csv', delimiter = ',')
df
```

```python
num_columns = df.select_dtypes('number').columns
num_columns = ['Year', 'Distance', 'Engine_capacity(cm3)', 'Price(euro)', 'km_year']
df_num = df[num_columns]
```

# Предобработка данных

Оценим количество данных:

```python
df_num.shape
```

Проверим на пропуски:

```python
df_num.isna().sum()
```

```python
df_num.hist(figsize=(25,20))
```

Разобъем датасе на признаки и целевую переменные:

```python
X,y = df_num.drop(columns = ['Price(euro)']).values, df_num['Price(euro)'].values
features_names = df_num.drop(columns = ['Price(euro)']).columns
```

Разделим данные на обучающую, валидационную и тестовую выборки:

```python
from train_test_spliter import train_test_split

x_train, y_train, x_val, y_val, x_test, y_test = train_test_split(X, y, train_size = 0.5, test_size=0.3, )    
print(f" Размер выборки для обучения {x_train.shape}, \n размер целевой переменной для обучения {y_train.shape}, \n размер выборки для валидации: {x_val.shape}, \n размер целевой переменной для валидации: {y_val.shape},  размер выборки для тестирования {x_test.shape}, размер целевой выборки для тестирования {y_test.shape}\n \n ")
```

Перед тем как проводить дальнейшие эксперименты оценим базовую модель:

```python
y_mean = y_test.mean()
y_hat = np.full(y_test.shape, y_mean)
y_hat.shape
```

```python
base_lin_model = LinearRegression(X=x_test)
base_mse = base_lin_model.score(X=x_test, y=y_test, yhat=y_hat)
base_mse 
```

```python
y_random_hat = np.random.randint(y_test.min(),y_test.max(),size=y_test.size)
y_random_hat.size
```

```python
base_random_lin_model = LinearRegression(X=x_test)
base_random_mse = base_random_lin_model.score(X=x_test, y=y_test, yhat=y_random_hat)
base_random_mse
```

```python
y_random_hat = y_test.mean()+y_test.std()*np.random.randn(*y_test.shape)
y_random_hat.size
```

```python
base_random_lin_model = LinearRegression(X=x_test)
base_random_mse = base_random_lin_model.score(X=x_test, yhat=y_random_hat, y=y_test)
base_random_mse
```

Приведем к единому масштабу:

```python
df_num.describe()
```

```python
from MinMaxScaler import MinMaxScaler
```

```python
scaler  = MinMaxScaler()

# Найдем минимальное и максимальное значение для каждого признака
scaler.fit(x_train)
print(f"Min value features:{scaler.min}")
print(f"Max value features:{scaler.max}")
```

```python
x_train_transformed = scaler.transform(x_train)
pd.DataFrame(x_train_transformed, columns=features_names).describe()
```

```python
# аналогично преобразуем валидационные и тестовую выборки
val_scaler = MinMaxScaler()
val_scaler.fit(x_val)
x_val_transformed = val_scaler.transform(x_val)

test_scaler = MinMaxScaler()
test_scaler.fit(x_test)
x_test_transformed = test_scaler.transform(x_test)
```

# Обучение модели

```python
regr = LinearRegression(X=x_train_transformed, learning_rate=0.5,epochs=300,batch_size=3000)
regr.fit(x_train_transformed, y_train)

regr.plot_cost()

print( 'train R2: %.4f; test R2: %.4f' % (regr.score(x_train_transformed, y_train), regr.score(x_test_transformed, y_test)))
```

посмотрим веса

```python
w, b = regr.get_w_and_b()

regr.vis_weigths(weights=np.squeeze(w), names=features_names)
```

```python
np.sum(np.abs(w))
```

### Эластичная регуляризация
Отметим, что во многих случаях неизвестно какая модель регуляризации окажется лучше, поэтому целесообразно использовать их обе вместе. Такая модель регуляризации называется эластической регуляризацией. Попробуем ее реализовать

```python
from ElasticReg import ElasticRegression

regr = ElasticRegression(X=x_train_transformed, learning_rate=0.5,epochs=300,batch_size=3000, l1_penalty = 0.001,l2_penalty = 0.001)
regr.fit(x_train_transformed, y_train)

regr.plot_cost()

print(regr.score(x_test_transformed, y_test), regr.score(x_train_transformed, y_train))
```

посмотрим веса

```python
w, b = regr.get_w_and_b()

regr.vis_weigths(w, features_names)
```

```python
np.sum(np.abs(w))
```

# Подбор гиперпараметров

```python
alpha = [0.5, 0.3, 0.1, 0.01]
```

```python
for i in range(0, len(alpha)):
    regr = ElasticRegression(X=x_train_transformed, learning_rate=alpha[i],epochs=300,batch_size=3000, l1_penalty = 0.001,l2_penalty = 0.001)
    regr.fit(x_train_transformed, y_train)

    print(f"Test score: {regr.score(x_test_transformed, y_test):.2f}, {regr.score(x_train_transformed, y_train):.2f}")
```
