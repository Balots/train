# Практическая работа 7. Bagging


**Бэггинг (от англ. bootstrap aggregating)** — это метод машинного обучения, который заключается в обучении множества моделей на различных подвыборках исходного набора данных и последующем их объединении для получения более точного и устойчивого результата.

Основные идеи бэггинга:

1. Создание подвыборок.

    Из исходного набора данных случайным образом с возвращением выбираются подвыборки. 
    
    Это означает, что некоторые объекты могут попадать в подвыборку несколько раз, а некоторые могут не попасть вовсе.

2. Обучение моделей. 

    На каждой подвыборке обучается отдельная модель. 
    
    В качестве базовых моделей могут использоваться любые алгоритмы машинного обучения, такие как деревья решений, линейные модели и т.д.

3. Объединение результатов. 

    Для получения итогового результата прогнозы всех моделей объединяются. 
    
    Для задач классификации обычно используется голосование, а для задач регрессии — усреднение.

**Преимущества бэггинга:**

- Уменьшение дисперсии.

    Бэггинг позволяет уменьшить дисперсию ошибки модели, что делает ее более устойчивой к выбросам и шуму в данных.

- Улучшение обобщающей способности.

    За счет обучения на разных подвыборках модели получают возможность улавливать различные закономерности в данных, что улучшает их способность обобщаться на новые данные.
    
- Простота реализации.

    Бэггинг достаточно прост в реализации и может быть применен к любым алгоритмам машинного обучения.

**Недостатки бэггинга:**

- Увеличение вычислительных затрат.

    Обучение множества моделей требует больше вычислительных ресурсов, чем обучение одной модели.

- Потеря интерпретируемости.

    Объединение результатов множества моделей может затруднить интерпретацию полученного результата.

### Пример на синтетических данных

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

# Создаем обучающую выборку
X,y = make_classification(n_samples=100, n_classes=2, n_clusters_per_class=2, n_features=2, class_sep=.8, random_state=3, n_redundant=0)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap, ListedColormap
import seaborn as sns

palette = sns.color_palette(n_colors=2)
cmap = ListedColormap(palette)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.8)
```

```python
# Разделяем выборку на обучающую и тестову
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"Размер обучающей выборки: {len(X_train)}-{len(y_train)}")
print(f"Размер тестовой выборки: {len(X_test)}-{len(y_test)}")
```

```python
from random import randint

# Функция для создания подборки
def create_subset(train_x: list, train_y: list, n: int = None):
    '''
    Input:
        train_x - список данных
        train_y - список меток
        n - количество объектов в итоговой выборке
    '''
    subset_x = list()
    subset_y = list()
    
    if n==None:
        n = len(train_x)
    
    for i in range(n):
        index = randint(0,len(train_x)-1)
        subset_x.append(train_x[index])
        subset_y.append(train_y[index])
    
    return [subset_x,subset_y]
```

```python
# Протестируем 
n=len(X_train)
subset_x, subset_y = create_subset(train_x=X_train, train_y=y_train, n=n)

print(f"Итоговая обучающая выборка: \n{subset_x[:5]}, \nметки {subset_y[:5]}")
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Функция обучения модели на нескольких подвыборках
def train_bags(bags_count: int, train_x: list, train_y: list, model) -> list:
    '''
        Input:
            bags_count - количество подвыборок
            train_x - данные для обучения
            train_y - целевая переменная
            model - классификатор 
    '''
    result_bags = list()
    for i in range(bags_count):
        current_model = model()
        new_train_x, new_train_y = create_subset(train_x,train_y)
        
        d_x = np.array(new_train_x)
        d_y = np.array(new_train_y)        
        
        try:
            current_model.fit(d_x,d_y)
            result_bags.append(current_model)
        except Exception as Error:
            print(f"Не удалось обучить модель. {Error}")
    
    return result_bags
```

```python
def predict_on_bags(bags_models: list, test_x: list, verbose: bool = True) -> list:
    '''
        Получаем предсказания на нескольких моделях. Усредняем результат предсказаний.
    '''
    all_predicts_y = np.zeros((len(bags_models),len(test_x)))
    
    if verbose:
        print(f"Кол-во объектов: {len(test_x)}")
        print(f"Предсказание моделей:")

    for i, model in enumerate(bags_models):
        try:
            predict_y = model.predict(test_x)
            if verbose:            
                print(f"{i} -{predict_y}")
            all_predicts_y[i,:] = predict_y
        except:
            print("Ошибка предсказания модели")
    
    if verbose:
        print(f"Все предсказания: \n {all_predicts_y}")
    
    # усредняем предсказанные значения
    return np.round(all_predicts_y.mean(axis=0))
```

```python
lin_model = LogisticRegression

bags_model = train_bags(bags_count=3, train_x=X_train, train_y=y_train, model=lin_model)
result_y = predict_on_bags(bags_models=bags_model, test_x=X_test)

print(f"Предсказанные значения модели Логистическая регрессия: {result_y}")
print(f"Метки классов: {y_test}")
```

```python
from sklearn.metrics import f1_score

f1_metrics = f1_score(y_true=y_test,y_pred=result_y)

print(f"Значение метрики f1 = {f1_metrics:.2f}")
```

```python
dt_model = DecisionTreeClassifier

bags_model = train_bags(bags_count=3, train_x=X_train, train_y=y_train, model=dt_model)
result_y = predict_on_bags(bags_models=bags_model, test_x=X_test)

print(f"Предсказанные значения модели Деревья решений: {result_y}")
print(f"Метки классов: {y_test}")
```

```python
from sklearn.metrics import f1_score

f1_metrics = f1_score(y_true=y_test,y_pred=result_y)

print(f"Значение метрики f1 = {f1_metrics:.2f}")
```

### Класс BaggingClassifier в Sklearn

<code>
from skelarn.ensemble import BaggingClassifier <br>
BaggingClassifier(estimator=None, n_estimators=10, *, max_samples=1.0, max_features=1.0, bootstrap=True, bootstrap_features=False, oob_score=False, n_jobs=None, random_state=None)
</code>

Параметры BaggingClassifier

- estimator: Базовая модель, которая будет использоваться для обучения на подвыборках. По умолчанию - DecisionTreeClassifier.
- n_estimators: Количество базовых моделей (классификаторов), которые будут созданы и обучены. Чем больше моделей, тем выше вероятность получить более точный результат, но и увеличивается время обучения.
- max_samples: Количество объектов в каждой подвыборке. Может быть целым числом (абсолютное количество) или числом с плавающей точкой от 0 до 1 (доля от размера исходной выборки). По умолчанию - 1.0 (т.е. размер подвыборки равен размеру исходной выборки).
- max_features: Количество признаков, которые будут использоваться для обучения каждой базовой модели. Аналогично max_samples, может быть целым числом или числом с плавающей точкой. По умолчанию - 1.0.
- bootstrap: Булевский параметр, указывающий, использовать ли бутстрап (случайный выбор с возвращением) при создании подвыборок. По умолчанию - True.
- bootstrap_features: Булевский параметр, указывающий, использовать ли бутстрап для выбора признаков для каждой базовой модели. По умолчанию - False.
- oob_score: Использовать ли вневыборочные (out-of-bag) наблюдения для оценки ошибки обобщения. Доступно только в том случае, если bootstrap=True. Out-of-bag наблюдения — это те объекты, которые не попали в конкретную подвыборку, на которой обучается данная модель.
- n_jobs: Количество параллельно выполняемых задач для обучения (fit) и предсказания (predict). None означает 1, если только не используется в контексте joblib.parallel_backend. -1 означает использование всех процессоров. Подробнее см. в "Глоссарии".
- random_state: Параметр, отвечающий за генерацию случайных чисел. Используется для воспроизводимости результатов.

### Класс RandomForestClassifier в Sklearn

from sklearn.ensemble import Random

RandomForestClassifier(n_estimators=100, *, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None, monotonic_cst=None)


Основные параметры: 
- n_estimators: (по умолчанию = 100) — 
Количество деревьев в лесу. Чем больше деревьев, тем лучше, как правило, но это увеличивает время обучения.
- criterion: (по умолчанию = "gini") — Функция измерения качества разделения. 
Может быть "gini" (индекс Джини) или "entropy" (энтропия) . 
- max_depth: (по умолчанию = None) — Максимальная глубина дерева. 
- min_samples_split: (по умолчанию = 2) — Минимальное количество выборок, необходимое для разделения внутреннего узла. 
Чем больше значение, тем меньше вероятность переобучения.
- min_samples_leaf: (по умолчанию = 1) — Минимальное количество выборок, которое должно быть в листовом узле. 
Аналогично min_samples_split, помогает предотвратить переобучение.
- max_features: (по умолчанию = "sqrt" или 1/3 для классификации) — Количество признаков, которые рассматриваются при выборе наилучшего разделения. 
Может быть целым числом, долей (float), <br>
"sqrt" (квадратный корень из числа признаков), <br> 
"log2" (логарифм по основанию 2 из числа признаков) или None (все признаки). 
<br> "sqrt" часто используется по умолчанию и хорошо работает. <br>
- bootstrap: (по умолчанию = True) — Использовать ли бутстрап (случайный выбор с возвращением) для обучения каждого дерева. 
Если False, используется весь набор данных. Обычно рекомендуется оставлять True.
- oob_score: (по умолчанию = False) — Использовать ли out-of-bag (OOB) выборки для оценки обобщающей способности. 
Если True, то после обучения можно получить OOB-оценку с помощью атрибута oob_score_.
- n_jobs: (по умолчанию = None) — Количество параллельно выполняемых задач. -1 означает использование всех процессоров. 
Полезно для ускорения обучения на многоядерных системах.
- random_state: (по умолчанию = None) — Генератор случайных чисел. 
Фиксирование random_state обеспечивает воспроизводимость результатов.
- class_weight: (по умолчанию = None) — Веса, связанные с классами, в форме {метка_класса: вес}. Если не задано, предполагается, что все классы имеют вес, равный единице. Для задач с несколькими выходами может быть предоставлен список словарей в том же порядке, что и столбцы y. Режим "balanced" использует значения y для автоматической корректировки весов обратно пропорционально частотам классов во входных данных как n_samples / (n_classes * np.bincount(y)). Режим "balanced_subsample" аналогичен "balanced", за исключением того, что веса вычисляются на основе выборки bootstrap для каждого выращенного дерева.

#### Пример на синтетике

```python
from collections import OrderedDict
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier

RANDOM_STATE = 123

# Сгенерируем данные
X, y = make_classification(
    n_samples=500,
    n_features=25,
    n_clusters_per_class=1,
    n_informative=15,
    random_state=RANDOM_STATE,
)

ensemble_clfs = [
    (
        "RandomForestClassifier, max_features='sqrt'",
        RandomForestClassifier(
            warm_start=True,
            oob_score=True,
            max_features="sqrt",
            random_state=RANDOM_STATE,
        ),
    ),
    (
        "RandomForestClassifier, max_features='log2'",
        RandomForestClassifier(
            warm_start=True,
            max_features="log2",
            oob_score=True,
            random_state=RANDOM_STATE,
        ),
    ),
    (
        "RandomForestClassifier, max_features=None",
        RandomForestClassifier(
            warm_start=True,
            max_features=None,
            oob_score=True,
            random_state=RANDOM_STATE,
        ),
    ),
]

# Создадим словарь с названием классификаторов
error_rate = OrderedDict((label, []) for label, _ in ensemble_clfs)

# Зададим диапазон количества деревьев
min_estimators = 15
max_estimators = 150

for label, clf in ensemble_clfs:
    for i in range(min_estimators, max_estimators + 1, 5):
        clf.set_params(n_estimators=i)
        clf.fit(X, y)

        # Оценим классификатор на данных которые не вошли в подвыборку
        oob_error = 1 - clf.oob_score_
        error_rate[label].append((i, oob_error))

# Построим график 
for label, clf_err in error_rate.items():
    xs, ys = zip(*clf_err)
    plt.plot(xs, ys, label=label)

plt.xlim(min_estimators, max_estimators)
plt.xlabel("Количество деревьев")
plt.ylabel("OOB ошибки")
plt.legend(loc="upper right")
plt.show()
```

### Класс RandomForestRegressor в Sklearn

from sklearn.ensemble import RandomForestRegressor

RandomForestRegressor(n_estimators=100, *, criterion='squared_error', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=1.0, max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, ccp_alpha=0.0, max_samples=None, monotonic_cst=None)


Основные параметры: 
- n_estimators: (по умолчанию = 100) — Количество деревьев в лесу. Чем больше деревьев, тем лучше, как правило, но это увеличивает время обучения.
- criterion: (по умолчанию = "mse") — Функция измерения качества разделения. Для регрессии обычно используется "mse" (mean - squared error — средняя квадратичная ошибка) или "mae" (mean absolute error — средняя абсолютная ошибка).
- max_depth: (по умолчанию = None) — Максимальная глубина дерева. Если None, то глубина не ограничена, что может привести к переобучению. Рекомендуется устанавливать разумное значение.
- min_samples_split: (по умолчанию = 2) — Минимальное количество выборок, необходимое для разделения внутреннего узла. Чем больше значение, тем меньше вероятность переобучения.
- min_samples_leaf: (по умолчанию = 1) — Минимальное количество выборок, которое должно быть в листовом узле. Аналогично - - min_samples_split, помогает предотвратить переобучение.
- max_features: (по умолчанию = "auto") — Количество признаков, которые рассматриваются при выборе наилучшего разделения. Может быть целым числом, долей (float), "sqrt" (квадратный корень из числа признаков), "log2" (логарифм по основанию 2 из числа признаков) или None (все признаки). "auto" или "sqrt" часто используются по умолчанию и хорошо работают.
- bootstrap: (по умолчанию = True) — Использовать ли бутстрап (случайный выбор с возвращением) для обучения каждого дерева. Если False, используется весь набор данных. Обычно рекомендуется оставлять True.
oob_score: (по умолчанию = False) — Использовать ли out-of-bag (OOB) выборки для оценки обобщающей способности. Если True, то после обучения можно получить OOB-оценку с помощью атрибута oob_score_.
- n_jobs: (по умолчанию = None) — Количество параллельно выполняемых задач. -1 означает использование всех процессоров. Полезно для ускорения обучения на многоядерных системах.

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold

X, y = fetch_california_housing(return_X_y=True, as_frame=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

regressor = RandomForestRegressor(min_samples_leaf=5, random_state=0, n_jobs=-1, oob_score=True)
regressor.fit(X_train, y_train)
# Выведем ошибку на данных не участвующие в обучении
print(regressor.oob_score_)
```
