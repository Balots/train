## **Метод figure**


<div class='alert alert-block alert-warning'>
При создания графика MatplotLib создает объекта класса Figure:

`plt.figure(num, figsize = None, layout = None)`

figure - это некий контейнер для хранения графиков.

Где:

* num - уникальный идентификатор полотна (по-умолчанию num = None). Позволяет редактировать уже созданный график.

* figsize - размер полотна в дюймах. По-умолчанию (6.4, 4)

* layout - режим размещения элементов внутри полотна ('constarined' - корректирует размер графика, чтобы вписать туда все элементы, 'compressed' - аналогично вписывает графики, убирая внутренние отступы)

Если figure не прописывать в программе, то объект создается с параметрами по умолчанию.
</div>


````python
import matplotlib.pyplot as plt

# Создадим пустой контейнер размером 5, 2.7 дюйма
plt.figure(figsize=(5, 2.7), layout='constrained')
````

## **Линейные графики**

</div>


<div class='alert alert-block alert-warning'>

Линейный график создается с помощью метода:

`plt.plot([x], y, [fmt])`

Основные параметры для функции plt.plot():
* x, y: Массивы данных для осей x и y. x не обязательный параметр, если его не задавать, создает массив последовательных чисел с 0 до количества элементов в y.
* color: Цвет линии (например, 'red', 'blue', 'green', или шестнадцатеричный код).
* linewidth: Толщина линии.
* linestyle: Стиль линии (возможные значения: '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted').
* marker: Маркер для точек данных (например, 'o', 'x', '+', 's').
* markersize: Размер маркера.
* label: Метка для легенды. Поддерживает встроенный редактор формул
* [fmt]: Формат маркер и стиль линии можно задать одной строкой (например 'o-')
</div>


````python
import numpy as np
import matplotlib.pyplot as plt

# Пример 2. Объект figure создается по умолчанию.
# Создаем 100 точек в диапазоне от 0 до 2 с равномерным распределением
y = np.linspace(0, 2, 100)

# Добавляем линейный график, значение x задается по умолчанию, толщину, маркер и стиль линии задаем c помощью параметров
plt.plot(y, color = 'red', linewidth = 0.1, linestyle = 'solid', marker='*')
````

## **Методы форматирования графиков**

</div>


Основные методы добавления элементов:
* plt.xlabel('text') - подпись оси x
* plt.ylabel('text') - подпись оси y
* plt.title('text') - заголовок графика
* plt.grid() - сетка
* plt.legend() - легенда:
  * loc: Указывает положение легенды на графике.
  
    Возможные значения:
    'best': Выбирает оптимальное расположение автоматически
    
    'upper right', 'upper left', 'lower left', 'lower right',
  
    'center left', 'center right', 'center',
    
    'upper center', 'lower center',  'outside'
  * bbox_to_anchor: Позволяет задать точное положение легенды в виде координат от 0 до 1.
  * title: Заголовок легенды.
  * frameon: Включает или отключает рамку вокруг легенды (True или False).
  * fontsize: Размер шрифта в легенде.

Основные методы форматирования графика:
* plt.xlim(минимум, максимум)  - ограничение значения оси x
* plt.ylim(минимум, максимум) - ограничение значения оси y
* plt.text(координата x, координата y, 'текст') - добавляет текст для заданной точки
* plt.xticks(), plt.xticks() - форматирование меток оси X, Y:
  * ticks: Список значений, соответствующих положениям меток на оси X.
  * labels: Список строковых меток, которые будут отображаться для каждого значения в ticks. Если не указан, будут использоваться значения из ticks.
  * kwargs: Дополнительные ключевые аргументы для форматирования меток, такие как rotation, fontsize, fontweight и другие.


````python
# Пример построения графика с форматирование меток оси X
import matplotlib.pyplot as plt
import numpy as np

# Создаем данные для графика
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Строим график, label - текст для легенды
plt.plot(x, y, label = 'sin(x)')

# Настраиваем метки на оси X
plt.xticks([0, 3.14, 6.28], ['0', 'π', '2π'], fontsize=14, color='red', rotation = 45)
plt.xlabel('Угол (рад)', fontsize=16)

# Добавляем сетку
plt.grid(True)
# заголовок
plt.title('Синус', fontsize=18)
# Добавляем легенду в нижний левый угол
plt.legend(loc='lower left')
# Отображаем график
plt.show()
````

````python
# Построим график сигмоиды.
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))

plt.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.legend()
plt.show()
````

````python
# Пример построения графика с форматрование легенды
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)

#Добавляем легенду
plt.legend(loc='center left', ncol=2, labels=['sin(x)', 'cos(x)'], title="Тригонометрические функции")
plt.show()
````

### **Метод save:**

<div class='alert alert-block alert-warning'>

Сохранить полученный рисунок с графиками можно с помощью метода savefig:

Параметры:

````python
# Пример №3
import math
import matplotlib.pyplot as plt
import numpy as np

# Пустое полотно можно сохранять в переменную и редактировать его в любом месте программы
fig = plt.figure(num=1, figsize=(9,3))
N = 10
# используем logspace чтобы задать логарифмеческое распределение от 0.1 до 1 для 10 точек. endpoint - включает последний элемент в набор данных
x1 = np.logspace(start = 0.1, stop = 1, num = 10, endpoint=True)
x2 = np.logspace(start = 0.1, stop = 1, num = 10, endpoint=False)
y = np.zeros(N)

plt.plot(x1, y, 'o')
plt.plot(x2, y+0.5, 'o-')

# Зададим ограничение для оси Y
plt.ylim([-0.5, 1])
# Добавим легенду не в область графика, а на полотно
fig.legend(["x1", "x2"], title="Log")
plt.title("Логарифмеческое распределение")
# Сохраним рисунок в файл logspace.jpg, с разрешением - 50
plt.savefig('logspace.jpg', dpi=50)
````

## **Класс Axes: несколько графиков на одном полотне и не только**

</div>


Класс Axes() управляет областями, где отображаются данные, включая оси, метки, линии и другие элементы графика. 
Он позволяет создавать несколько подграфиков на одном полотне.

Создавать и получать доступ к областям axes внутри контейнера можно несколькими способами. Рассмотрим основные:





<div class='alert alert-block alert-warning'>

### **Метод subplots:**


</div>


<div class='alert alert-block alert-warning'>
plt.subplots() cоздает нескольких графиков на одной фигуре. Он позволяет организовать данные  в сетку, что особенно полезно при сравнении различных наборов данных или при отображении многомерных данных.


```
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows, ncols, index)
```

Параметры:

* nrows: Количество строк в сетке.
* ncols: Количество столбцов в сетке.
* index: Позиция подграфика в сетке, нумерация начинается с 1 в верхнем левом углу и идет слева направо и сверху вниз.

Возвращаемые значения (необязательно):
* fig - контейнер с графиками
* axes - объект Axes или массив объектов Axes, в зависимости от количества созданных координатных осей.

</div>

````python
import matplotlib.pyplot as plt

# по-умолчанию создает фигуру с одним Axes (графиком)
fig, ax = plt.subplots()
````

Обращаться к объектам axes можно по их индексу в массиве или в методе subplot

````python
import numpy as np
import matplotlib.pyplot as plt

# Создаем данные
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))
# Создаем фигуру с двумя подграфиками в одной строке. Обращаемся по index к первому графику
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Синус')

# Обращаемся ко второму графику
plt.subplot(1, 2, 2)
plt.plot(x, y2, color = 'red')
plt.title('Косинус')

plt.show()
````

Для того чтобы добавить отступы между графиками на сетке воспользуемся методом  tight_layout

```
matplotlib.pyplot.tight_layout(*, pad=1.08, h_pad=None, w_pad=None, rect=None)
```

* pad:float, по-умолчанию =1.08. Отступ между краем рисунка и краями подграфиков, в долях от размера шрифта.

* h_pad, w_pad: float, default: pad. Отступ (высота/ширина) между краями смежных подграфиков, как доля размера шрифта.


````python
import matplotlib.pyplot as plt
import numpy as np

# Создаем фигуру с двумя подграфиками. Сохраняем в переменные fig и axes контейнер и массив подграфиков
fig, axes = plt.subplots(nrows=2, ncols=1)

# Строим данные на первом подграфике
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
axes[0].plot(x, y1)
# Добавим два графика на второй координатной оси.
y2 = np.cos(x)
y3 = np.cos(x*x)
axes[1].plot(x, y2)
axes[1].plot(x, y3)
#добавим отступы между графиками
plt.tight_layout(pad=1.08, h_pad=5, w_pad=0)
plt.show()
````

### **Класс Axes**

</div>

Объекты класса Axes можно добавлять отдельно внутри контейнера.


Например если создать объекта класса через одноименный метод plt.axes и указать координаты графика, он создаст координатную ось и добавит её в определенной позиции внутри полотна

```
plt.axes(rect, **kwargs)
```

Параметры:
* rect: list, tuple. Список с координатами точки и размером графика
* **kwargs: другие параметры характерные для класса Axes

````python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes((0.1, 0.1, 0.8, 0.8), facecolor='yellow')  # [left, bottom, width, height]
ax = plt.axes([0.65, 0.6, 0.2, 0.2], facecolor='grey')  # [left, bottom, width, height]
ax = plt.axes([0.2, 0.6, 0.2, 0.2], facecolor='cyan')  # [left, bottom, width, height]
````

* объект класс Axes содержит множество встроенных методов. Например ранее рассмотренный метод plot, работает аналогично и для объектов axes

### *Метод plt.pie:*

</div>


<div class='alert alert-block alert-warning'>

Функция **plt.pie** в библиотеке Matplotlib позволяет создавать круговые диаграммы. Данные разбиваются на категории (сегменты) и отображается площадь каждого сегмента, соответствующую его доле в общем объеме.

```
plt.pie(x, explode=None, labels=None, colors=None, autopct=None, shadow=False, startangle=0, radius=1,wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False)

```

Основные параметры
* x: Последовательность чисел, представляющая доли или абсолютные значения для каждого сектора круговой диаграммы.
* labels: Последовательность строк, соответствующих меткам для каждого сектора.
* colors: Последовательность цветов для каждого сектора.
* explode: Последовательность чисел, указывающая, на сколько долей радиуса будет выдвинут каждый сектор.
* startangle: Угол в градусах, с которого начинается первый сектор.
* autopct: Формат для отображения процентов внутри каждого сектора.

Дополнительные параметры:

* shadow: Булево значение, определяющее, будет ли у диаграммы тень.
* radius: Радиус круговой диаграммы.
* wedgeprops: Словарь с дополнительными свойствами для каждого сектора (например, linewidth, edgecolor).
* textprops: Словарь с дополнительными свойствами для текста меток и процентов.
* center: Координаты центра круговой диаграммы.
* frame: Булево значение, определяющее, будет ли вокруг диаграммы рамка.
* rotatelabels: Булево значение, определяющее, будут ли метки повернуты вдоль окружности.
</div>

### **Гистограммы**




Данный тип диаграммы позволяет визуализировать распределение данных, разделяя их на интервалы (bins) и подсчитывая количество точек данных, попадающих в каждый интервал.


<div class='alert alert-block alert-warning'>

Распределения данных в виде гистограммы в pyplot можно построить с помощью метода hist или plot(kind='bar').


```
n, bins, patches = plt.hist(data, bins=None, range=None, density=False, weights=None, cumulative=False, **kwargs)
```

Основные параметры функции plt.hist():

* **data**: Массив или список значений, для которых нужно построить гистограмму.
* Необязательные:
  * **bins**:  Число интервалов (bins) на оси x. Можно указать: Целое число: будет создано указанное количество интервалов с равными шириной.
Массив: будут использованы заданные границы интервалов.
Функция: будет использована для динамического определения границ интервалов.

  * **range**: Диапазон значений данных, для которых будет построена гистограмма.
Указывается в виде списка из двух элементов: [начало, конец].

  * **density**: Если True, вместо частоты (количества точек данных) на оси y будет отображаться плотность распределения (нормированная на единицу).

  * **weights**: Массив весов для каждой точки данных.
Вес влияет на вклад точки данных в расчет частоты/плотности в каждом интервале.
cumulative: (по умолчанию False) Если True, будет построена накопительная гистограмма.
  * **histtype**: Тип гистограммы: 'bar' (стандартная), 'step', 'stepfilled'.
  * **align**: Выравнивание баров: 'mid', 'left', 'right'.
  * **orientation**: Ориентация гистограммы: 'vertical' или 'horizontal'.
  * **color**: Цвет баров.
  * **edgecolor**: Цвет границ баров.
  * **label**: Метка для легенды.

Возвращаемые значения:

* n: Массив, содержащий количество элементов в каждом интервале (бинне). Это фактически высоты столбцов гистограммы.
* bins: Массив, содержащий границы интервалов. Он на один элемент длиннее, чем массив n, так как включает в себя левую границу первого интервала и правую границу последнего.
* patches: Список объектов Patch, представляющих каждый столбец гистограммы. Эти объекты можно использовать для дальнейшей настройки внешнего вида столбцов, например, изменения цвета, ширины или прозрачности.
</div>

````python
import matplotlib.pyplot as plt
import numpy as np

# Данные
data = np.random.randn(1000)

# Построим гистограмму для 20 интервалов, зададим границы черным цветом
plt.hist(data, bins=20, edgecolor='black')

# Настройка заголовка и подписей осей
plt.title('Гистограмма')
plt.xlabel('Значение')
plt.ylabel('Частота')

# Отображение сетки
plt.grid(True)

# Отображение графика
plt.show()
````

````python
# Добавление нескольких типов графиков на одну систему координат
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10**7)
mu = 121
sigma = 21
x = mu + sigma * np.random.randn(1000)

num_bins = 100

n, bins, patches = plt.hist(x, num_bins,
                            density = 1,
                            color ='green',
                            alpha = 0.7)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

plt.plot(bins, y, '--', color ='black')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.title('Гистограмма распределения данных\n\n',
          fontweight = "bold")

plt.show()
````

### **Метод boxplot**


</div>


Диаграмма boxplot, также известная как диаграмма ящиков, коробчатые диаграммы или диаграмма «ящик с усами», представляет собой график, который наглядно демонстрирует распределение и сравнение между собой значений 25-го, 50-го, 75-го процентилей, а также минимальных, максимальных и выбросов набора данных.


Снимок экрана 2024-08-07 112418.png



```
matplotlib.pyplot.boxplot(data, notch=None, vert=None, patch_artist=None, widths=None)
```

Основные параметры
* data: Может быть массивом, списком или DataFrame. Определяет данные, которые будут использованы для построения ящиковых диаграмм.

Необязательные параметры:
* notch: bool. Если True, то на ящиках будут отображены вырезы, которые указывают на доверительный интервал для медианы.
* vert: bool. Если True (по умолчанию), ящики будут ориентированы вертикально. Если False, то горизонтально.
* patch_artist: bool. Если True, то ящики будут заполнены цветом.
* widths: float or array(float). Определяет ширину ящиков.
* labels: list[str]. Используется для подписи ящиков.
* boxpropsdict: dict. Задаёт внешний вид боксам
* medianpropsdict: dict. Задаёт внешний вид медианному значению
* whis: float or (float, float). Задаёт размер усов. По-умолчанию = 1.5 межквартального размаха.

````python
import matplotlib.pyplot as plt
import numpy as np

# Создадими массив случайных данных из 100 примеро в 3 признаками
data = np.random.randn(100, 3)
print(f"Размер данных {data.shape}")
# Строим ящиковые диаграммы
plt.boxplot(data, patch_artist=True, notch=True, vert=False, medianprops={'color': 'yellow', 'linewidth': 2.3}, labels=['Группа 1', 'Группа 2', 'Группа 3'],
            whis=1.5, flierprops={'marker': 'o', 'markerfacecolor': 'red'})
# Пояснение кода:
# patch_artist=True: Ящики будут заполнены цветом.
# notch=True: На ящиках будут отображены вырезы.
# vert=False: Ящики будут ориентированы горизонтально.
# labels=['Группа 1', 'Группа 2', 'Группа 3']: Ящики будут подписаны.
# whis=1.5: Усы будут простираться на 1.5 межквартильного размаха.
# medianprops={'color': 'yellow', 'linewidth': 2.3}: Линия медианны будет желтой.
# flierprops={'marker': 'o', 'markerfacecolor': 'red'}: Выбросы будут обозначены красными кружками.

plt.title('Ящиковые диаграммы')
plt.xlabel('Группы')
plt.ylabel('Значения')

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
# создадим массив синтетических данных с нормальным распределением из 37 примеров с 4 признаками
data = np.random.lognormal(size=(37, 4), mean=1.5, sigma=1.75)

# сохраним параметры в отдельные переменные
labels = list('ABCD')
fs = 10  # fontsize
#настроим отображения для одних и тех же данных по-разному
boxprops = dict(linestyle='--', linewidth=3, color='darkgoldenrod')
medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
flierprops = dict(marker='o', markerfacecolor='green', markersize=12, markeredgecolor='none')

meanpointprops = dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick')
meanlineprops = dict(linestyle='--', linewidth=2.5, color='purple')

# создадим сетку из 2 строк с 3 столбцами
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(6, 6), sharey=True)
axs[0, 0].boxplot(data, boxprops=boxprops)
axs[0, 0].set_title('Custom boxprops', fontsize=fs)

axs[0, 1].boxplot(data, flierprops=flierprops, medianprops=medianprops)
axs[0, 1].set_title('Custom medianprops\nand flierprops', fontsize=fs)
# увеличим
axs[0, 2].boxplot(data, whis=(0, 100))
axs[0, 2].set_title('whis=(0, 100)', fontsize=fs)

axs[1, 0].boxplot(data, meanprops=meanpointprops, meanline=False, showmeans=True)
axs[1, 0].set_title('Custom mean\nas point', fontsize=fs)

axs[1, 1].boxplot(data, meanprops=meanlineprops, meanline=True, showmeans=True)
axs[1, 1].set_title('Custom mean\nas line', fontsize=fs)

axs[1, 2].boxplot(data, whis=[15, 85])
axs[1, 2].set_title('whis=[15, 85]\n#percentiles', fontsize=fs)

for ax in axs.flat:
    ax.set_yscale('log')
    ax.set_yticklabels([])

fig.suptitle("I never said they'd be pretty")
fig.subplots_adjust(hspace=0.4)
plt.tight_layout()
plt.show()
````

### **Метод violinplot**

plt.violinplot() — это функция в библиотеке Matplotlib, которая используется для создания скрипичных диаграмм. Эти диаграммы представляют собой более информативный вариант ящиковых диаграмм, поскольку они отображают плотность распределения данных в каждой точке.

```
matplotlib.pyplot.violinplot(data, vert=True, widths=0.5, showmeans=False, showextrema=True, showmedians=False, quantiles=None, points=100, bw_method=None, side='both', *, data=None)
```

Основные параметры
* data: Обязательный параметр. Может быть массивом, списком или DataFrame. Определяет данные, которые будут использованы для построения скрипичных диаграмм.
* vert: Булевый параметр. Если True (по умолчанию), скрипки будут ориентированы вертикально. Если False, то горизонтально.
* widths: Число или массив чисел. Определяет ширину скрипок.
* showmeans: Булевый параметр. Если True, то на графике будут отображены средние значения для каждой группы.
* showmedians: Булевый параметр. Если True, то на графике будут отображены медианы для каждой группы.

    ax = seaborn.histplot(data, x, y=None, hue=None, stat='count', bins='auto', cumulative=False, common_norm=True, multiple='layer', element='bars', fill=True, kde=False, palette=None, legend=True, ax=None, **kwargs)

**Обязательные параметры:**
* data: Данные для построения гистограммы. Это может быть pandas DataFrame, NumPy array или любой другой объект, который можно преобразовать в эти форматы.

**Основные параметры для настройки:**

* element: Визуальный элемент, который будет использоваться для отображения гистограммы: 'bars' (обычные столбцы), 'step' (ступенчатая линия).
* fill: Если False, отображает только контуры гистограммы.
* palette: Цветовая палитра для отображения категорий.
* legend: Если True, отображает легенду.

**Другие полезные параметры:**

### pairplot: Матрица графиков рассеяния для всех пар числовых переменных


    seaborn.pairplot(data, *, hue=None, palette=None, vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1)

**Обязательные параметры:**
* data: DataFrame, содержащий данные для построения графиков.

**Основные параметры для настройки:**
* hue: Имя столбца, по которому будут группироваться данные. Каждый уровень категории будет отображен своим цветом.
* vars: Список имен столбцов, которые необходимо включить в графики. По умолчанию используются все числовые столбцы.

**Другие полезные параметры:**

### barplot: Столбчатые диаграммы

    seaborn.barplot(data=None, x=None, y=None, hue=None, order=None, estimator='mean',  errorbar=('ci', 95), palette=None)

Основные параметры и их описание:
* x, y: str. Имена столбцов в DataFrame, которые будут использованы для осей x и y соответственно. Обычно ось x содержит категориальные данные, а ось y - числовые.
* data: DataFrame. Данные, которые будут использованы для построения графика. Обычно это Pandas DataFrame.
* hue: str. Имя столбца, который будет использоваться для разделения данных на категории и отображения их с помощью разных цветов.
* palette: строка или словарь. Палитра цветов для отображения категорий (если используется параметр 'hue').
* order: список. Порядок, в котором категории будут отображены на оси x.
* estimator: функция (по умолчанию np.mean). Функция, которая будет использоваться для агрегирования данных по каждой категории. По умолчанию используется среднее значение.
* orient: str (по умолчанию 'v'). Ориентация столбцов: 'v' (по умолчанию) или 'h'.
* errorbar: str, (string, number), tuple, callable or None. Имя метода errorbar (либо «ci», «pi», «se» или «sd»), либо кортеж с именем метода и параметром уровня, либо функция, которая сопоставляет вектор с интервалом (min, max), либо None, чтобы скрыть errorbar.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Выберем датасет c информацие про чаевые
df_tips = sns.load_dataset('tips')
# Построим распределение чаевых в зависимости от пола.
# Функция - сумма, errorbar=None - отключим
ax = sns.barplot(data=df_tips, x="tip", y="sex", estimator='sum', errorbar=None)

# добавим подписи для столбцов
ax.bar_label(ax.containers[0], fontsize=10)
ax.set_xlabel("Чаевые")
ax.set_ylabel("Пол")
ax.set_title("Размер чаевых")
````

seaborn.countplot - это функция в библиотеке Seaborn, предназначенная для создания столбчатых диаграмм, которые отображают частоту появления различных категорий в данных.

* data: DataFrame. Данные, которые будут использованы для построения графика. Обычно это Pandas DataFrame.

### catplot: Более гибкий вариант для категориальных данных

    seaborn.catplot(data=None, x=None, y=None, hue=None, row=None, col=None, kind='strip', estimator='mean', color=None, palette=None, legend='auto', sharex=True, sharey=True, margin_titles=False, facet_kws=None)

Catplot в Seaborn — это универсальная функция для создания различных видов графиков, основанных на категориальных данных. Она объединяет в себе несколько других функций, таких как barplot, boxplot, violinplot и др., позволяя легко переключаться между ними.

**Основные параметры и их значения**
* data: DataFrame. Исходный набор данных, содержащий категориальные и числовые переменные.
* x, y: Строка. Имена столбцов в DataFrame, соответствующих категориальной и числовой переменным соответственно. Определяют оси графика.
hue: Строка. Имя столбца в DataFrame, содержащего категориальную переменную, по которой будут разделены данные на группы. Каждая группа будет представлена своим цветом или маркером.
* kind: str
  Тип графика:
    * 'strip': Точечная диаграмма
    * 'swarm': Точечная диаграмма с избеганием наложения точек
    * 'box': Ящичковая диаграмма
    * 'violin': Скрипичная диаграмма
    * 'boxen': Модифицированная ящичковая диаграмма
    * 'kdeplot': Диаграмма плотности ядра
    * 'point': Точечная диаграмма со средним значением и доверительным интервалом
    * 'bar': Столбчатая диаграмма
* estimator: Строка. Функция группировки данных по интервалам. По-умолчанию mean.
* order, hue_order: Список строк. Порядок отображения категорий по осям x и для параметра hue соответственно.
* orient: Строка ('v' или 'h'). Ориентация графика: вертикальная или горизонтальная.
* color: Цвет (строка или число). Цвет для всех элементов графика.
* palette: Строка, список или словарь. Палитра цветов для разделения данных по группам, заданным параметром hue.
* height, aspect: Число. Высота и пропорции фигуры графика.
* col, row: Для создания многогранных графиков.
* col_wrap: Максимальное количество столбцов в многогранном графике.
* margin_titles: Добавление заголовков к графикам в многогранном графике.
* sharex, sharey: Общие оси для всех графиков в многогранном графике.
* legend: Управление легендой графика.

## Express vs Graph Objects

Plotly предоставляет два основных интерфейса для создания графиков:
* Graph Objects: позволяет создавать сложные графики, с индивидуальной настройкой
* Express: упрощенный вариант создания распространенных типов графиков, предоставляет удобный интерфейс для задания данных и основных параметров.

Когда использовать какой подход?
* Graph Objects:
  * Для создания сложных и кастомизированных графиков
  * Когда требуется полный контроль над каждым элементом графика
  * Для реализации специфических визуализаций, не поддерживаемых Express
* Express:
  * Для быстрого создания распространенных типов графиков
  * Для интерактивной визуализации данных
  * Когда приоритетом является простота использования

````python
import plotly.express as px

# загрузим набор данных
df = px.data.tips()

# методы px возвращают объект который можно потом использовать
fig = px.scatter(df, x="total_bill", y="tip", color="sex")

fig.update_layout(
    title=dict(text='Настройка шрифтов', x=0.5),
    font_family="Courier New",
    font_color="blue",
    title_font_family="Times New Roman",
    title_font_color="red",
    legend_title_font_color="green"
)
fig.update_xaxes(title_font_family="Arial")
fig.show()
````

## Plotly.GlobalObject

Класс позволяющий управлять графиком по частям.

Основные компоненты graph_objects:

* Figure: Корневой объект, представляющий весь график.
* Trace: Отдельный набор данных, который отображается на графике. Каждый след может иметь свой тип (линия, точки, гистограмма и т.д.).
* Layout: Определяет внешний вид графика в целом, включая заголовки, оси, легенды и другие элементы.

````python
import plotly.graph_objects as go

# Создаем первый график распределения
trace1 = go.Scatter(
    x=[1, 2, 3],
    y=[4, 5, 6],
    mode='lines+markers'
)

# Создаем второй график распределения
trace2 = go.Scatter(
    x=[2, 3, 4],
    y=[4, 3, 2],
    mode='lines+markers'
)

# Создаем фигуру и добавляем графики
fig = go.Figure(data=[trace1, trace2])

# Настраиваем форматирование
fig.update_layout(
    title="Пример графика",
    xaxis_title="Ось X",
    yaxis_title="Ось Y"
)

fig.show()
````

Ключевые параметры в graph_objects:

Figure:

* data: Список следов.
* layout: Словарь с параметрами компоновки.
* frames: Для создания анимаций.

Trace:
* x, y: Данные для оси X и Y.
* mode: Тип следа (линия, точки, гистограмма и т.д.).
* name: Имя следа для легенды.
* marker: Параметры маркеров (цвет, размер, форма).
* line: Параметры линии (цвет, ширина, стиль).

Layout:

* title: Заголовок графика.
* xaxis, yaxis: Параметры осей.
* legend: Параметры легенды.
* margin: Отступы графика.
* colorway: Цветовая палитра.
... и многие другие параметры

````python
# Сравнение создание простого графика в Express и Graph Objects
import pandas as pd
# Plotly Express
import plotly.express as px
# Graph Objects
import plotly.graph_objects as go


df = pd.DataFrame({
  "Fruit": ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"],
  "Contestant": ["Alex", "Alex", "Alex", "Jordan", "Jordan", "Jordan"],
  "Number Eaten": [2, 1, 3, 1, 3, 2],
})

fig = px.bar(df, x="Fruit", y="Number Eaten", color="Contestant", barmode="group")
fig.show()

fig = go.Figure()
for contestant, group in df.groupby("Contestant"):
    fig.add_trace(go.Bar(x=group["Fruit"], y=group["Number Eaten"], name=contestant,
      hovertemplate="Contestant=%s<br>Fruit=%%{x}<br>Number Eaten=%%{y}<extra></extra>"% contestant))
fig.update_layout(legend_title_text = "Contestant")
fig.update_xaxes(title_text="Fruit")
fig.update_yaxes(title_text="Number Eaten")
fig.show()
````

## Метод figure_factory

figure_factory  - это модуль, предоставляющий набор функций для создания специфических типов графиков, которые могут быть сложны или громоздки для создания с помощью обычных графических объектов Plotly.

````python
import plotly.figure_factory as ff
import numpy as np

# Создадим данные
x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 2
x4 = np.random.randn(200) + 4

# Объединим
hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Создадим график распределения данных с размером бинов = 2
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
````

Основные типы графиков, создаваемых с помощью figure_factory:

* Матрицы корреляции: create_corr
* Гистограммы: create_distplot
* 3D поверхности: create_surface
