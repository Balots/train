```python
def f1(x):
    return np.sin(x)

def f2(x):
    return np.cos(x)

def f3(x):
    return np.cos(x) * np.cos(x)

x = np.linspace(-10, 10, 10000)
y1, y2, y3 = f1(x), f2(x), f3(x)

plt.plot(x, y1, 'r')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Гармонические функции')
plt.savefig('4quest.png')
```

```python
x = np.linspace(0, 15, 10)
y1 = np.exp(x)
y2 = np.array([i**2 for i in x])

plt.plot(x, y1, linestyle='--', marker='o', color='black', label=r'$e^{x}$')
plt.plot(x, y2, linestyle='dotted', marker='s', color='green', label=r'$x^2$')

plt.legend()
plt.savefig('quest 5.png')
```

```python
loss_model_A = np.array([0.66514, 0.67245, 0.66087,0.65276,0.65822,0.65949,0.66138,0.65429,0.64371,0.65075,0.64019])
precision_model_A = np.array([0.89964, 0.90657, 0.91591, 0.91605, 0.92104, 0.91529, 0.90995, 0.90184, 0.90429,0.91541,0.90867])
recall_model_A = np.array([0.74877, 0.75123, 0.75117, 0.75123, 0.74704, 0.74513, 0.74668, 0.74673, 0.74472, 0.7463,  0.74631])
                     
fig, axs = plt.subplots(3, 1, figsize=(10, 6))

x = np.arange(11) 

axs[0].plot(x, loss_model_A, label='loss of model A')
axs[0].set_title('Loss')
axs[0].set_xlabel('index')
axs[0].set_ylabel('weights')
axs[0].legend()


axs[1].plot(x, precision_model_A, label='presicion of model A')
axs[1].set_title('Presicion')
axs[1].set_xlabel('index')
axs[1].set_ylabel('weights')
axs[1].legend()

axs[2].plot(x, recall_model_A, label='recall of model A')
axs[2].set_title('Recall')
axs[2].set_xlabel('index')
axs[2].set_ylabel('weights')
axs[2].legend()


plt.tight_layout(pad=1.08, h_pad=5, w_pad=0)
```

```python
df = pd.read_csv('results.csv', sep=',')
loss_model = df['train/dfl_loss']
precision_model = df['metrics/precision(B)']
recall_model = df['metrics/recall(B)']
x = df['epoch']

fig, axs = plt.subplots(3, 1, figsize=(10, 6))

axs[0].plot(x, loss_model, label='loss of model')
axs[0].set_title('Loss')
axs[0].set_xlabel('index')
axs[0].set_ylabel('weights')
axs[0].legend()


axs[1].plot(x, precision_model, label='presicion of model')
axs[1].set_title('Presicion')
axs[1].set_xlabel('index')
axs[1].set_ylabel('weights')
axs[1].legend()

axs[2].plot(x, recall_model, label='recall of model')
axs[2].set_title('Recall')
axs[2].set_xlabel('index')
axs[2].set_ylabel('weights')
axs[2].legend()


plt.tight_layout(pad=1.08, h_pad=5, w_pad=0)
```
