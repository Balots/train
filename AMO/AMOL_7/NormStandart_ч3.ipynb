# Способы масштабирования свойств

```python
import pandas as pd

data =[
[1000, 10,0.5],
[765,5,0.35],
[800,7,0.09]]

df = pd.DataFrame(data, columns = ['A','B','C'])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize =(9, 4))

for i in enumerate(df.columns):    
    axs[i[0]].boxplot(df[i[1]])
```

```python
# Распределение данных
import seaborn as sns

sns.kdeplot(df.A)
plt.show()
sns.kdeplot(df.B)
plt.show()
sns.kdeplot(df.C)
plt.show()
```

## Нормализация

```python
# Нормализуем данные
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x = min_max_scaler.fit_transform(df)
df_scaled = pd.DataFrame(x, columns=df.columns)
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize =(9, 4))

for i in enumerate(df.columns):            
    axs[i[0]].boxplot(df_scaled.iloc[:,i[0]])
    axs[i[0]].set_title(i[1])
```

## Стандартизация данных

```python
# Нормализуем данные
from sklearn import preprocessing

standart_scaler = preprocessing.StandardScaler()
x = standart_scaler.fit_transform(df)
df_scaled = pd.DataFrame(x, columns=df.columns)
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize =(9, 4))

for i in enumerate(df.columns):    
    axs[i[0]].boxplot(df_scaled[i[1]])
    axs[i[0]].set_title(i[1])
```

```python
robust_scaler = preprocessing.RobustScaler()
x = robust_scaler.fit_transform(df)
df_scaled = pd.DataFrame(x, columns=df.columns)
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize =(9, 4))

for i in enumerate(df.columns):    
    axs[i[0]].boxplot(df_scaled[i[1]])
    axs[i[0]].set_title(i[1])
```
