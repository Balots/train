```python
def rgb2gray(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    gray_value = 0.29 * r + 0.59 * g + 0.11 * b
    gray_value = gray_value.astype(np.uint8)
    return np.array([gray_value.T]*3).T
# отобразим исходное изображение
plt.subplot(2,2,1)
plt.imshow(image)
plt.axis('off')
# конвертируем исходное изображение в градации серого
converted_image = rgb2gray(image)
# отобразим новое изображение
plt.subplot(2,2,2)
plt.imshow(converted_image)
plt.axis('off')
```

```python
def rgb2gray_average(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    gray_value = np.mean(image, axis = 0) * 0.5 + np.mean(image, axis = 0) * 0.5
    gray_value = gray_value.astype(np.uint8)
    return np.array([gray_value.T]*3).T
# отобразим исходное изображение
plt.subplot(2,2,1)
plt.imshow(image)
plt.axis('off')
# конвертируем исходное изображение в градации серого первым способом
converted_image = rgb2gray(image)
# отобразим новое изображение
plt.subplot(2,2,2)
plt.imshow(converted_image)
plt.axis('off')
# конвертируем исходное изображение в градации серого вторым способом
average_image = rgb2gray_average(image)
# отобразим новое изображение
plt.subplot(2,2,3)
plt.imshow(average_image )
plt.axis('off')
```
